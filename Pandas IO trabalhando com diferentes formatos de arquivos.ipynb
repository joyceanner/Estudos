***Desafio : lendo dados com vários parâmetros***




*   Verifique se o arquivo CSV está separado por vírgula ou ponto e vírgula.
*   A codificação do arquivo é ISO-8859-1.
*   As três primeiras linhas linhas do arquivo podem ser desconsideradas, pois *   o cabeçalho só começa na quarta linha.
*   As 9 últimas linhas também podem ser desconsideradas, pois são apenas informações sobre onde os dados foram obtidos.
*   Para deletar as últimas linhas é necessário adicionar o parâmetro engine='python'.

CRIANDO BANCO DE DADOS

```python
engine = create_engine('sqlite:///:memory:')
```

# **Escrevendo em banco de dados local**

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'
```

```python
dados = pd.read_csv(url)
```

```python
dados.head()
```

# **Colocando essas informações dentro do banco de dados**

```python
dados.to_sql('Clientes', engine, index=False)
```

```python
inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```

# **Fazendo consultas no banco de dados**

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renda="Empregado" '
```

```python
empregados = pd.read_sql(query, engine)
```

```python
empregados
```

```python
empregados.to_sql('Empregados', con=engine, index=False)
```

```python
pd.read_sql_table('empregados', engine)
```

```python
pd.read_sql_table('empregados', engine, columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual'])
```

# **Como atualizar um banco de dados**

```python
query = 'SELECT * FROM clientes'
```

```python
pd.read_sql(query,engine )
```

```python
query = 'DELETE FROM clientes WHERE ID_Cliente = 5008804'
with engine.connect() as conn:
  conn.execute(query)
```

```python
pd.read_sql_table('clientes', engine)
```

Atualizando informações

```python
query = 'UPDATE clientes SET Grau_escolaridade="Ensino Superior" WHERE ID_cliente=5008808'
with engine.connect() as conn:
  conn.execute(query)
```

```python
pd.read_sql_table('clientes', engine)
```
