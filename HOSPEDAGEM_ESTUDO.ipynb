## ENTENDENDO O PROBLEMA


Importando o Pandas

```python
import pandas as pd
```

Lendo o arquivo JSON

```python
dados = pd.read_json('/content/dados_hospedagem.json')
dados.head()
```

Os dados estão aninhados em dicionários, será aplicado a normalização desses dados.

```python
dados = pd.json_normalize(dados['info_moveis'])
dados
```

## DADOS NÚMERICOS

Nele, vemos dados numéricos, de texto e também listas com os dois tipos.



Verificando as colunas do dataframe

```python
colunas = list(dados.columns)
colunas
```

Para realizar o tratamento focado em cada tipo de dado, precisaremos realizar um desagrupamento, removendo-os das listas. Faremos isso a seguir.

```python
dados = dados.explode(colunas[3:])
dados
```

Removendo o Index

```python
dados.reset_index(inplace = True, drop =  True)
dados.head()
```

Verificando o tipo de casa coluna

```python
dados.info()
```

Importando numpy

```python
import numpy as np
```

Transformando a coluna 'MAX_HOSPEDES' em número

```python
dados['max_hospedes'] = dados['max_hospedes'].astype(np.int64)
```

Confirmando a alteração

```python
dados.info()
```

Separando 3 colunas numa variavel e transformando em número

```python
col_numericas = ['quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']
```

```python
 dados[col_numericas] = dados[col_numericas].astype(np.int64)
```

Transformando a coluna Avaliação geral em float

```python
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(np.float64)
```

Confirmando as alterações

```python
dados.info()
```

Substituindo $ e ',' por espaços vazios, utilizando o strip para retirar os espaços vazios do inicio e do fim

```python
dados['preco'] = dados['preco'].apply(lambda x: x.replace('$', '').replace(',', '').strip())
```

Transformando em float

```python
dados['preco'] = dados['preco'].astype(np.float64)
```

```python
dados.info()
```

## DADOS TEXTUAIS

```python
dados[['taxa_deposito' , 'taxa_limpeza']] = dados[['taxa_deposito' , 'taxa_limpeza']].applymap((lambda x: x.replace('$', '').replace(',', '').strip()))
```

```python
dados[['taxa_deposito' , 'taxa_limpeza']] = dados[['taxa_deposito' , 'taxa_limpeza']].astype(np.float64)
```

```python
dados.info()
```

Transformando as strings da coluna em minusculos

```python
dados['descricao_local'] = dados['descricao_local'].str.lower()
```

```python
dados.head()
```

Visualizando a coluna "Descrição Local"

```python
dados['descricao_local'][3169]
```

Retirando todos os caracteres que não são necessários para a analise de dados ( Usado regex para separar os caracteres que serão permitidos " letras, números, - e '.

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
```

Retirando os "-" que não estão dentro de uma palavra composta

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w) - (?!\w)', ' ', regex=True)
```

Retirando os espaços em branco no inicio e fim

```python
dados['descricao_local'] = dados['descricao_local'].str.split()
dados.head()
```

Retirando os caracteres "{ , }" da coluna Comodidades

```python
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|"', '', regex=True)
```

Criando uma lista de token colocando o split() colocando como parâmetro o nosso caractere separador, as vírgulas, entre aspas simples

```python
dados['comodidades'] = dados['comodidades'].str.split(',')
dados.head()
```

Fazendo o processo de tokenização para a coluna descricao_vizinhanca presente no conjunto de dados

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('(?<!\w) - (?!\w)', ' ', regex=True)
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.lower()
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.split()
dados.head()
```
